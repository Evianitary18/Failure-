```python
import requests
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup as bs4

# Ambil halaman
url = 'https://simple.wikipedia.org/wiki/List_of_countries_by_continents'
continents_page = requests.get(url).text

# Parsing HTML
continents_countries_soup = bs4(continents_page, 'lxml')

# Cari elemen <span> dengan class mw-headline di dalam <h2>
continents = continents_countries_soup.find_all('span', {"class": "mw-headline"})

# Ekstrak nama benua dari elemen
continent_names = [continent.text for continent in continents]

# Tampilkan hasil
print(continent_names)
```
